```python
import pandas as pd
from sqlalchemy import create_engine , text
```

```python
username = 'postgres'
password = 'hellosql'
database = 'Amazon'
host = 'localhost'
port = '5432' #default PostgreSQL port

# Create the connection string
connection_string = f'postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}'
        
# Create the engine
engine = create_engine(connection_string)
```

```python
amazon_sql = 'SELECT * FROM amazon_prime_titles'
uncleaned_df = pd.read_sql(amazon_sql, engine)
#setting the streaming platform to amazon prime for all titles
uncleaned_df['streaming_platform'] = 'Amazon Prime'
uncleaned_df
```

```python
# creating a dataframe with just movies

query = "SELECT show_id, type,title,date_added, release_year,rating, duration, listed_in FROM amazon_prime_titles WHERE type = 'Movie';"
movies_df = pd.read_sql(query, engine)
movies_df['streaming_platform'] = 'Amazon Prime'
movies_df
```

```python
movies_df.isna().sum()
#date_added for the majority of the rows are null so it will be dropped
#rating is null for a small minority of the data so those rows will be dropped
#also dropping show_id as it is not needed for our data analysis
```

```python
movies_df = movies_df.drop(columns=['date_added','show_id'])
movies_df = movies_df.dropna()
movies_df
```

```python
movies_df.duplicated().sum() # no duplicated rows in the dataframe
```

```python
#looking for Erroneous Values

print(movies_df['release_year'].value_counts())
print(movies_df['rating'].value_counts())
print(movies_df['duration'].value_counts())
print(movies_df['listed_in'].value_counts())
```

```python
#replacing Erroneous Values with values in line with the rest of the data for ratings
movies_df['rating'] = movies_df['rating'].replace('AGES_18_', '18+')
movies_df['rating'] = movies_df['rating'].replace('AGES_16_', '16+')
movies_df['rating'] = movies_df['rating'].replace('ALL_AGES', 'ALL')
movies_df['rating'] = movies_df['rating'].replace('16', '16+')
movies_df['rating'] = movies_df['rating'].replace('NOT_RATE', 'NR')
movies_df['rating'].value_counts()
```

```python
movies_df.info()  #looking at the data types and which ones to change    
```

```python
#creating a new column for duration to create an int value for minutes
def slice_minutes(min):
    return int(min[:-4])

movies_df['duration_minutes'] = movies_df['duration'].apply(slice_minutes)
movies_df.info()
```

```python
#dropping the original duration column as it is no longer needed
movies_df = movies_df.drop(columns='duration')
movies_df.info()
```

```python
# creating a dataframe with just TV shows
query = "SELECT show_id, type,title,date_added, release_year,rating, duration, listed_in FROM amazon_prime_titles WHERE type = 'TV Show';"
tv_df = pd.read_sql(query, engine)
tv_df['streaming_platform'] = 'Amazon Prime'
tv_df
```

```python
tv_df.isna().sum()  
#date_added is mostly null so the column will be dropped
# rating nulls are a small minority so those rows will be dropped
#also dropping show_id as it is not needed for our data analysis
```

```python
tv_df = tv_df.drop(columns=['date_added', 'show_id'])
tv_df = tv_df.dropna()
tv_df
```

```python
tv_df.duplicated().sum() # no duplicated rows in the dataframe
```

```python
#looking for Erroneous Values

print(tv_df['release_year'].value_counts())
print(tv_df['rating'].value_counts())
print(tv_df['duration'].value_counts())
print(tv_df['listed_in'].value_counts())
```

```python
tv_df.info()
```

```python
#creating a new column for duration to have an int value for seasons
def slice_seasons(season):
    if len(season) <= 8:
        return int(season[:-7])
    else:
        return int(season[:-8])

    
tv_df['duration_seasons'] = tv_df['duration'].apply(slice_seasons)

tv_df.info()
```

```python
#dropping the original duration column
tv_df = tv_df.drop(columns='duration')
tv_df.info()
```

```python
#creating new csv files for each dataframe to use with the rest of our data
outgoing_movies_df = movies_df.to_csv('AmazonMovies.csv', index = True)
outgoing_tv_df = tv_df.to_csv('AmazonTvShows.csv', index = True)
```
